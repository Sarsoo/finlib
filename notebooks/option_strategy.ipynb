```python
import pyfinlib
import matplotlib.pyplot as plt
```

```python
# strategy_components = pyfinlib.options.strategy.bull_call_spread(100, 10, 110, 10)
# strategy_components = pyfinlib.options.strategy.iron_butterfly_spread(100, 10, 110, 10)
strategy_components = pyfinlib.options.strategy.long_straddle(100, 10, 10)
# strategy_components = pyfinlib.options.strategy.long_strangle(100, 10, 110, 10)
# strategy_components = pyfinlib.options.strategy.iron_butterfly_spread(110, 10, 100, 10)
strategy = pyfinlib.Strategy()
for i in strategy_components:
    strategy.add_component(i)
```

```python
ran = range(50, 150)
x = [i for i in ran]
payoff = [strategy.payoff(i) for i in ran]
profit = [strategy.profit(i) for i in ran]
```

```python
# Plot the historical returns and VaR threshold
plt.figure(figsize=(10, 6))
plt.plot(x, payoff, linestyle='dashed', label='Payoff')
plt.plot(x, profit, label='Profit')
plt.fill_between(x, profit, where=[i > 0 for i in profit], interpolate=True, color="green", alpha=0.2)
plt.fill_between(x, profit, where=[i < 0 for i in profit], interpolate=True, color="red", alpha=0.2)
plt.title('Profit')
plt.xlabel('Underlying Price')
plt.ylabel('Profit')
plt.legend()
plt.grid()
plt.show()
```
